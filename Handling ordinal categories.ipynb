### Handling ordinal categories

```python
import pandas as pd
import datetime
```

```python
date_base=datetime.datetime.today()
date_list=[date_base-datetime.timedelta(days=x)for x in range(0,20)]
df=pd.DataFrame(date_list)
df.columns=['Days']
df
```

```python
df['day_of_week']=df['Days'].dt.day_name()
df.head()
```

```python
week_map={'Monday':1,
         'Tuesday':2,
         'Wednesday':3,
         'Thursday':4,
         'Friday':5,
         'Saturday':6,
         'Sunday':7}
```

```python
df['ordinal_day']=df['day_of_week'].map(week_map)
df
```
